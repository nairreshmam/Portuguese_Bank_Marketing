##### Description: This represents the number of times a customer was contacted during previous campaigns.
##### Conclusion:
A higher number of contacts may indicate higher persistence, but an increase in contacts without success (i.e., "no" outcome) could indicate diminishing returns.
If many contacts were needed in the past but still no subscription occurred, the likelihood of a subscription now may be lower.
##### 3. campaign (Number of Contacts in Current Campaign)
##### Description: This refers to the number of times a customer has been contacted in the current campaign.
##### Conclusion:
Higher numbers in this column could signify persistence. However, diminishing returns may be present here too: after multiple contacts without success, the likelihood of success may drop.
If only a few contacts are made and a "yes" is achieved, it could indicate higher receptiveness from customers to the campaign.
##### 4. pdays (Days Since Last Contact in a Previous Campaign)
##### Description: This shows the number of days since the client was last contacted in a previous campaign. A value of -1 means the client was not previously contacted.
##### Conclusion:
Low values (few days since last contact) may indicate a strong likelihood of subscription if past engagements were positive.
A high pdays value could mean the customer has not been contacted for a long time, which may decrease the likelihood of them remembering or being interested.
If pdays = -1 (no previous contact), then the client is fresh, and we would expect the current campaign to play a larger role in determining success.
##### Key Insights:
Poutcome and pdays can be strong indicators of the likelihood of success based on past interactions.
Previous and campaign show how much effort has been put into contacting customers, but excessive contacts without success may indicate a decreasing likelihood of conversion

## Overall Conclusion for the analysis and visualization.
The dataset reveals that past interactions, such as the outcome of previous campaigns (poutcome) and days since last contact (pdays), are key predictors of whether a customer will subscribe to a bank product. Customers with successful past experiences or recent contact are more likely to say "yes." Additionally, fewer contacts in the current campaign (campaign), higher education, and white-collar jobs are associated with a higher likelihood of subscription. Customers with higher balances are also more inclined to subscribe, while excessive contact attempts and prior failures reduce the chances of success.
